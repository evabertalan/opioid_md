```python
import os
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def save_figure(fig, target):
    fig.savefig(target)
```

```python
PDB_CODE = '6b73'
SEQ = 'B'
folder = '../'+PDB_CODE+SEQ+'/'
results_folder = folder+'results/'
plot_folder = results_folder+'plots/'
```

### x-y plane over time

```python
xst_files = sorted([results_folder+'namd/'+i for i in os.listdir(results_folder+'namd/') if re.match(r'(step7.).*\_production.xst$', i)])
print(xst_files)
```

```python
box_inf = np.empty((0, 4))
for file in xst_files:
    dat = np.loadtxt(file,
                    usecols=[0,1,5,9],
                    dtype=np.float, 
                    comments='#')
    box_inf = np.concatenate([box_inf, dat])
```

```python
print(box_inf.shape)
```

```python
time = box_inf[:,0]
area = box_inf[:,1] * box_inf[:,2]
z = box_inf[:,3] 
volume = box_inf[:,1] * box_inf[:,2] * box_inf[:,3]

fig, ax = plt.subplots(1, figsize=(18,5))
ax.plot(time, area, linewidth=0.5, label='x-y plane area')
# ax.plot(time, z, linewidth=0.5, label='z axis')
# ax.plot(time, volume, linewidth=0.5, label='volume')

ax.title('Box parameters over the simulation')
ax.legend()
plt.show()
```

```python
save_figure(fig, plot_folder+'x-y_plane_over_simulation')
```
